```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
d=pd.read_excel("plz.xlsx")
d
```

```python
def ex(x, a, b, c, d):
    return a+b*np.log(c*x+d)

def lg(x,a,b,c,d):
    return a+b*np.log(c*x+d)
```

```python
popt, pcov = curve_fit(lg, d.t, d.dl,  method="lm")
popt, pcov
```

```python
plt.plot(d.t, d.dl)
plt.plot(d.t, lg(d.t,*popt))
```
