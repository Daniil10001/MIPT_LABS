```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
du=pd.read_excel("dataup.xlsx")
dd=pd.read_excel("datadown.xlsx")
du=du[np.isnan(du.P)==False]
dd=dd[np.isnan(dd.P)==False]
du
```

```python
def func(x, a, b):
    return a * np.exp(-b / x)
```

```python
Td=np.array(dd['T'])
Pd=np.array(dd['P'])
Tu=np.array(du['T'])
Pu=np.array(du['P'])
Ta=np.append(Tu,Td)
Pa=np.append(Pu,Pd)
```

```python
Ta,Pa
```

```python
popt, pcov = curve_fit(func, Ta, Pa, sigma=1, bounds=([1e6,0],[1e9,10000]))
```

```python
popt, pcov
```

```python
plt.scatter(Tu,Pu,marker='^',label="Повышение температуры")
plt.scatter(Td,Pd,marker='v',label="Понижение температуры")
plt.plot(Ta,func(Ta,*popt),'black')

plt.xlabel('Температура, К')
plt.ylabel('Давление, услов. мм')

plt.minorticks_on()
plt.grid(which='major')
plt.grid(which='minor', linestyle=':')
plt.legend()
plt.ylim(ymin=plt.gca().get_yticks()[0])
plt.xlim(xmin=plt.gca().get_xticks()[0])
plt.savefig('cringe.eps')
```

```python
def getkb(X,Y):
    k=(np.average(Y*X)-np.average(Y)*np.average(X))/(np.average(X*X)-np.average(X)**2)
    b=np.average(Y)-k*np.average(X)
    def av(n):
        return np.average(n)
    sigk=np.sqrt(np.abs((av(Y*Y)-av(Y)*av(Y))/(av(X*X)-av(X)**2) - k*k))/np.sqrt(X.shape[0])
    #print((av(Y*Y)-av(Y)*av(Y))/(av(X*X)-av(X)**2) - k*k)
    sigb=sigk*np.sqrt(av(X*X)-av(X)**2)
    return np.array([k,b,np.abs(sigk/k),np.abs(sigb/b)])
```

```python
Xa=1/Ta
Ya=np.log(Pa/np.max(Pa))
Xu=1/Tu
Yu=np.log(Pu/np.max(Pa))
Xd=1/Td
Yd=np.log(Pd/np.max(Pa))
```

```python
a,b,sa,sb=getkb(Xa,Ya)
```

```python
plt.scatter(Xu*1000,Yu,marker='^',label="Повышение температуры")
plt.scatter(Xd*1000,Yd,marker='v',label="Понижение температуры")
plt.plot(Xa*1000,Xa*a+b, 'black')

plt.xlabel(r'Обратная температура, $10^{-3}\frac{1}{К}$')
plt.ylabel(r'Логрифм давления, $P/P_{max}$')

plt.minorticks_on()
plt.grid(which='major')
plt.grid(which='minor', linestyle=':')
plt.legend()
plt.ylim(ymin=plt.gca().get_yticks()[0])
plt.xlim(xmin=plt.gca().get_xticks()[0])
plt.savefig('norm.eps')
```
