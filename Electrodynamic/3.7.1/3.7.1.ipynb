```python
def getkb(X,Y):
    k=(np.average(Y*X)-np.average(Y)*np.average(X))/(np.average(X*X)-np.average(X)**2)
    b=np.average(Y)-k*np.average(X)
    def av(n):
        return np.average(n)
    sigk=np.sqrt(np.abs((av(Y*Y)-av(Y)*av(Y))/(av(X*X)-av(X)**2) - k*k))/np.sqrt(X.shape[0])
    #print((av(Y*Y)-av(Y)*av(Y))/(av(X*X)-av(X)**2) - k*k)
    sigb=sigk*np.sqrt(av(X*X)-av(X)**2)
    return np.array([k,b,np.abs(sigk/k),np.abs(sigb/b)])
```

```python
X=nu1**2
Y=(V1/(nu1*I1))**-2
dY=2*((1e-3/V1)**2+(3e-6/I1)**2)**0.5
a1,b1,sa1,sb1=getkb(X,Y)
a1,b1,sa1,sb1, (1/b1)**0.5
```

```python
fig,ax = plt.subplots(figsize=(9,6))
ax.errorbar(X,Y,fmt='.', yerr=Y*dY)
ax.plot(X,X*a1+b1)
ax.minorticks_on()
ax.grid(which='major')
ax.grid(which='minor', linestyle=':')
ax.set_xlabel(r"$\nu^2$, $^{2}$")
ax.set_ylabel(r"$1/\xi^2$, $^{-2}$")
fig.savefig("ksi1.eps")
```

```python
def getk(x, y):
    a = np.sum(x * y) / np.sum(x**2)
    sigma_a = np.sqrt(np.sum((y - a * x)**2) / (x.shape[0] - 1) / np.sum(x**2))/a
    return a, sigma_a
```
