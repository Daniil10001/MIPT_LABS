```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getkb(X,Y):
    k=(np.average(Y*X)-np.average(Y)*np.average(X))/(np.average(X*X)-np.average(X)**2)
    b=np.average(Y)-k*np.average(X)
    def av(n):
        return np.average(n)
    sigk=np.sqrt(np.abs((av(Y*Y)-av(Y)*av(Y))/(av(X*X)-av(X)**2) - k*k))/np.sqrt(X.shape[0])
    #print((av(Y*Y)-av(Y)*av(Y))/(av(X*X)-av(X)**2) - k*k)
    sigb=sigk*np.sqrt(av(X*X)-av(X)**2)
    return np.array([k,b,np.abs(sigk/k),np.abs(sigb/b)])
```

```python
def IUp(x):
    return np.interp(x,d.Iu,d.Fu)/(72e-4)*1e-3

def IDn(x):
    return np.interp(x,d.Id[::-1],d.Fd[::-1])/(72e-4)*1e-3
```

```python
for i in range(1,5):
    fig,ax=plt.subplots()
    d1=pd.read_excel("data.xlsx",sheet_name=i)
    ax.plot(IUp(d1.Iu)**2,(d1.mu-d1.mu[0])*9.81/1000,"^")
    ax.plot(IDn(d1.Id)**2,(d1.md-d1.mu[0])*9.81/1000,"v")
    X=np.append(IUp(d1.Iu)**2,IDn(d1.Id)**2)
    Y=np.append(d1.mu-d1.mu[0],d1.md-d1.mu[0])*9.81/1000
    ax.set_title(Name[i])
    ax.set_xlabel("$B^2$, Тс$^2$")
    ax.set_ylabel("$F$, Н")
    ax.minorticks_on()
    ax.grid(which='major')
    ax.grid(which='minor', linestyle=':')
    a,b,sa,sb=getkb(X,Y)
    print(a,b,sa,sb)
    print(a*2*1.2566e-6/S[i])
    ax.plot(X,a*X+b)
    fig.savefig(f"{i}.eps")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Исходные данные
d1=pd.read_excel("data.xlsx",sheet_name=1)
ax.plot(IUp(d1.Iu)**2,d1.mu,"^")
ax.plot(IDn(d1.Id)**2,d1.md,"v")
y=np.append(IUp(d1.Iu)**2,IDn(d1.Id)**2)
x=np.append(d1.mu,d1.md)

# Добавляем константу (свободный член)
X = sm.add_constant(x)  # столбец единиц + x

model = sm.OLS(y, X).fit()
print(model.summary())

# Получаем предсказание и доверительный интервал
predictions = model.get_prediction(X)
ci = predictions.conf_int(alpha=0.05)  # 95% доверительный интервал

# Визуализация
plt.scatter(x, y, label='Данные')
plt.plot(x, model.fittedvalues, 'r-', label='Линия регрессии')
plt.fill_between(x, ci[:, 0], ci[:, 1], color='red', alpha=0.2, label='95% доверительная область')
plt.legend()
plt.show()
```
