```python
data=pd.read_csv('data_of_cmb.csv',header=None,names=['sort','N','factor','T, c'])
l=len(set(data['factor']))
F=np.zeros(l)
A=np.zeros(l)
m=list(set(data['factor']))
for i in range(len(m)):
    #print(i,m[i])
    F[i]=m[i]
    av=0
    for b in zip(data[data.factor==m[i]]['N'],data[data.factor==m[i]]['T, c']):
        av+=b[1]/b[0]**2
    A[i]=av/len(data.factor==m[i])
plt.scatter(F,A)
plt.ylabel(r'$O(n^2)$ factor')
plt.xlabel(r'comb sort factor')
plt.savefig('comb_n2_fc.png')
```

```python
plt.scatter(F,A)
plt.ylabel(r'$O(n^2)$ factor')
plt.xlabel(r'comb sort factor')
plt.xlim([1.2,1.35])
plt.ylim([0,0.025*1e-12])
plt.savefig('comb_n2_fcsc.png')
```

```python
data=pd.read_csv('data_of_cmb.csv',header=None,names=['sort','N','factor','T, c'])
l=len(set(data['factor']))
F=np.zeros(l)
A=np.zeros(l)
mF,mA=0,1
m=list(set(data['factor']))
for i in range(len(m)):
    #print(i,m[i])
    F[i]=m[i]
    av=0
    for b in zip(data[data.factor==m[i]]['N'],data[data.factor==m[i]]['T, c']):
        av+=b[1]/(b[0]*np.log(b[0]))
    A[i]=av/len(data.factor==m[i])
    if mA>A[i]:
        mA=A[i]
        mF=F[i]
plt.scatter(F,A)
plt.ylabel(r'$O(n \log n)$ factor')
plt.xlabel(r'comb sort factor')
plt.savefig('comb_logn_fc.png')
print(mF,mA)
```
