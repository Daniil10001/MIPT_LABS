```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data_of_sort_equal.csv',header=None)
data
```

```python
dt={}
for nm in set(data[0]):
    dt[nm]={'N':[],'T':[]}
for i in range(data.shape[0]):
    #dt.append({'index':i,data[0][i]:{'N':data[1][i], 'T':data[2][i]}})
    dt[data[0][i]]['N'].append(data[1][i])
    dt[data[0][i]]['T'].append(data[2][i])
for nm in set(data[0]):
    dt[nm]=pd.DataFrame(dt[nm])
dt
```

```python
datan=pd.concat(dt, axis=1,names=["sort", "data"])
datan
```

```python
data=pd.read_csv('data_of_sort_stan_2.csv',header=None,names=['sort','N','T, c'])

n2sort=['buble_sort','choose_sort','comb_sort','put_sort','shaker_sort']
data=(data.query("sort in ['buble_sort','choose_sort','comb_sort','put_sort','shaker_sort']"))
data['N']=np.log(data['N'])
data['T, c']=np.log(data['T, c'])
g = sns.FacetGrid(data, col="sort",col_wrap=3, margin_titles=True)
g.map(sns.regplot, "N", "T, c", color=".3",order=1,ci=99, marker="x", line_kws=dict(color="r"))
g.set(xlabel='Log N', ylabel='Log T')
```

```python
data=pd.read_csv('data_of_sort.csv',header=None,names=['sort','N','T, c'])
n2sort=['buble_sort','choose_sort','comb_sort','put_sort','shaker_sort']
data=(data.query("sort not in ['buble_sort','choose_sort','comb_sort','put_sort','shaker_sort']"))
data['T, c']=data['T, c']/data['N']
data['N']=np.log(data['N'])
g = sns.FacetGrid(data, col="sort",col_wrap=3, margin_titles=True)
g.map(sns.regplot, "N", "T, c", color=".3",order=1,ci=99, marker="x", line_kws=dict(color="r"))
g.set(xlabel='Log N', ylabel='T/N')
```

```python
n2sort=['buble_sort','choose_sort','comb_sort','put_sort','shaker_sort']
def get_dt(s):
    data__=pd.read_csv(s,header=None,names=['sort','N','T, c'])
    #data__=(data__.query('sort == "shaker_sort"'))#"sort not in ['buble_sort','choose_sort','comb_sort','put_sort','shaker_sort']"
    return data__

dat_s=get_dt('data_of_sort_sorted.csv')
dat_r=get_dt('data_of_sort_reverse.csv')
dat_e=get_dt('data_of_sort_stan_2.csv')
```

```python
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots(2,4, sharey=True)
ax=ax.reshape(8)
sns.set()
st=['buble_sort', 'choose_sort', 'comb_sort', 'heap_sort', 'merge_sort', 'put_sort', 'quick_sort', 'shaker_sort']
for i in range(8):
    sns.lineplot(dat_s[dat_s.sort==st[i]],x='N',y='T, c', ax=ax[i], label='sorted')
    sns.lineplot(dat_r[dat_r.sort==st[i]],x='N',y='T, c', ax=ax[i], label='reversed')
    sns.lineplot(dat_e[dat_e.sort==st[i]],x='N',y='T, c', ax=ax[i], label='random')
#plt.legend()
```
