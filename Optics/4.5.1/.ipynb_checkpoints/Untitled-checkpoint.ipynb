```python
Xp=np.array([])
Yp=np.array([])
for c in dp.columns:
    #print(dp[c].dropna())
    Xp=np.append(Xp,[c]*(dp[c].dropna().shape[0]))
    Yp=np.append(Yp,np.array(dp[c].dropna()))
Xp,Yp
```

```python
from scipy.optimize import curve_fit
def func(x, a, b, c,d):
    return a * np.sin(x*np.pi/180)**2+d
```

```python
popt, pcov = curve_fit(func, Xp, Yp)
popt
```
